**Dispersión Geográfica Tienda 1**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Supongamos que ya tienes tu DataFrame 'tienda_completa_1'
# Agrupar por vendedor y calcular la desviación estándar de latitud y longitud
dispersion_geografica_t1 = tienda_completa_1.groupby('Vendedor')[['lat', 'lon']].std().reset_index()

# Sumar la desviación estándar de latitud y longitud
dispersion_geografica_t1['Dispersion Total'] = dispersion_geografica_t1[['lat', 'lon']].sum(axis=1)
print(dispersion_geografica_t1)

# Calcular la Dispersión geográfica promedio
dispersion_geo_prom_t1 = round(np.mean(dispersion_geografica_t1['Dispersion Total']), 2)
print("Dispersión promedio Tienda 1:", dispersion_geo_prom_t1)

# Graficar la dispersión geográfica
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tienda_completa_1, x='lon', y='lat', hue='Vendedor', alpha=0.6)

# Añadir círculos que representen la desviación total
for i in range(dispersion_geografica_t1.shape[0]):
    plt.scatter(
        dispersion_geografica_t1['lon'][i],
        dispersion_geografica_t1['lat'][i],
        s=dispersion_geografica_t1['Dispersion Total'][i] * 100,  # Escalar el tamaño
        edgecolor='black',
        alpha=0.3,
        label=dispersion_geografica_t1['Vendedor'][i]
    )

plt.title('Dispersión Geográfica Tienda 1: Productos por Vendedor')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Leyenda ajustada
plt.grid()
plt.show()
```
